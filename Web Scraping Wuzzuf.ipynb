```python
#importing libraries
import pandas as pd
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
#importing the url

url='https://wuzzuf.net/search/jobs/?q=ilustrator&a=hpb'
```

```python
#create connetion
client= urlopen(url)
```

```python
#Getting the html code of the full page

html=client.read()
```

```python
html
```

```python
#closing the Request
client.close()
```

```python
#Create an HTML Parser Using BeautifulSoup

soup=bs(html,"html.parser")
soup
```

```python
#Create a Container for the Needed Date by class: css-1gatmva e1v1l3u10

containers=soup.find_all("div",{"class":"css-1gatmva e1v1l3u10"})
```

```python
len(containers)
```

```python
bs.prettify(containers[0])
```

```python
#Access Page Elements

containers[0].div.h2.text
```

```python
#get Jop Title
jopTitle=containers[0].find_all("h2",{"class":"css-m604qf"})
jopTitle[0].text
```

```python
#get Company Name

companyName=containers[0].find_all("a",{"class":"css-17s97q8"})
companyName[0].text
```

```python
#get Address

address=containers[0].find_all("span",{"class":"css-5wys0k"})
address[0].text
```

```python
#get jopType
jopType=containers[0].find_all("span",{"class":"css-1ve4b75 eoyjyou0"})
jopType[0].text
```

```python
#Bring it all Together

file2=open("Wuzzuf.csv","w")
```

```python
header="jop_title,Company_Name,jop_Type,Address,City,Country\n"
file2.write(header)
```

```python
for i in containers :
    jopTitle=i.find_all("h2",{"class":"css-m604qf"})
    jop_title=jopTitle[0].text.strip()

    companyName=i.find_all("a",{"class":"css-17s97q8"})
    Company_Name=companyName[0].text.strip()

    address=i.find_all("span",{"class":"css-5wys0k"})
    Address=address[0].text.strip()

    jopType=i.find_all("span",{"class":"css-1ve4b75 eoyjyou0"})
    jop_Type=jopType[0].text.strip()


    # print(jop_title)
    # print(Company_Name)
    #print(Address)
    #print(jop_title+", "+Company_Name+", "+Address+", "+jop_Type+ "/n")

    file2.write(jop_title + ", " +Company_Name+ ", " +jop_Type+ ", " +Address+ "\n")
file2.close()
```
